### 3. Linear Regression

```python
# Use the fit function and the score function in linear regression module of sklearn to  fit the data and observe the fit
# Fit the linear regression on mpg for the other features cylinder, horsepower, year and displacement
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression

linear_model = LinearRegression()
y = np.array(data['mpg'])
```

```python
# Simple linear regression with predictor cylinders and response mpg
print('Simple linear regression with predictor cylinders and response mpg')

x_c = np.array(data['cylinders']).reshape(-1,1)
simple_model_c = linear_model.fit(x_c,y)

print('The intercept is ' + str(simple_model_c.intercept_)) 
print('The slope is ' + str(simple_model_c.coef_))         
print('R-squared score is ' + str(simple_model_c.score(x_c,y)))
print('p-value is' + str(f_regression(x_c,y)[1]))

import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Simple linear regression with predictor displacement and response mpg
print('Simple linear regression with predictor displacement and response mpg')

x_d = np.array(data['displacement']).reshape(-1,1)
simple_model_d = linear_model.fit(x_d,y)

print('The intercept is ' + str(simple_model_d.intercept_))
print('The slope is ' + str(simple_model_d.coef_))
print('R-squared score is ' + str(simple_model_d.score(x_d,y)))
print('p-value is' + str(f_regression(x_d,y)[1]))
```

```python
# Simple linear regression with predictor horsepower and response mpg
print('Simple linear regression with predictor horsepower and response mpg')

x_h = np.array(data['horsepower']).reshape(-1,1)
simple_model_h = linear_model.fit(x_h,y)

print('The intercept is ' + str(simple_model_h.intercept_))
print('The slope is ' + str(simple_model_h.coef_))
print('R-squared score is ' + str(simple_model_h.score(x_h,y)))
print('p-value is' + str(f_regression(x_h,y)[1]))
```

```python
# Simple linear regression with predictor year and response mpg
print('Simple linear regression with predictor year and response mpg')

x_y = np.array(data['year']).reshape(-1,1)
simple_model_y = linear_model.fit(x_y,y)

print('The intercept is ' + str(simple_model_y.intercept_))
print('The slope is ' + str(simple_model_y.coef_))
print('R-squared score is ' + str(simple_model_y.score(x_y,y)))
print('p-value is' + str(f_regression(x_y,y)[1]))
```

*Problem 5.3 answer:*
 * All predictors appear to have a statistically significant relationship to the outcome as all p-values are very small ( 0.05).
 * The R-squared values are: 0.6482294003193044 (displacement) < 0.6059482578894348 (horsepower) < 0.6046889889441246 (cylinders) < 0.3370278133096227 (year).

```python
# Multiple Linear Regression with predictor cylinders, displacement, horsepower, weight, acceleration, year, origin and response mpg
print('Multiple Linear Regression with predictors cylinders, displacement, horsepower, weight, acceleration, year, origin and response mpg')

X = data[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']]
multi_model = linear_model.fit(X,y)

print('The intercept is ' + str(multi_model.intercept_))
print('The coefficent are ' + str(multi_model.coef_))
print('R-squared score is ' + str(multi_model.score(X,y)))

# Print the predictions of the your fitted model and the weights of Linear Regression (look at the function get_params())
y_pred = multi_model.predict(X)
print('Prediction of fitted model about mpg\n' + str(y_pred))
```

*Problem 5.4 answer:*

 (a) 
 The model fit better than all 4 simple linear regression models as R-squared value is 0.8214780764810599.
 
 (b)
   - Based on the below p-values, predictors displacement, weight, year, and origin have significant relationship ( 0.05) , while cylinders, horsepower, and acceleration do not (> 0.05).
   - The positive sign indicates that if the predictor increases, the response also increases. The negative sign indicates that if the predictor increases, the response decreases. For example: Predicate origin has positive sign of the coefficient with reference to response mpg. Origin in row 388 is 2, but then decreases to 1 in row 389, 390, 391. We can see that mpg in row 388 is 44.0, but then decreases to 32.0, 28.0, 31.0 in row 389, 390, 391. Example for negative sign of coefficient is the relationship between row 0, 1, 2, 3, 4 vs row 387, 388, 389, 390, 391 of predictor cylinders and response mpg.

```python
# Create a 2D matrix with each column representing all features
# Look up np.concatenate or np.stack
import statsmodels.api as sm
from statsmodels.formula.api import ols

multi_model = ols('mpg ~ cylinders + displacement + horsepower + weight + acceleration + year + origin', data=data).fit()
print(multi_model.summary())
```

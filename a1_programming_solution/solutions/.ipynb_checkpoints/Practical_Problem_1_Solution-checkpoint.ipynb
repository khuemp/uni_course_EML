### 2. Correlation

```python
feat1 = data['cylinders']
feat2 = data['mpg']
feat3 = data['displacement']
feat4 = data['horsepower']
feat5 = data['acceleration']
feat6 = data['year']
feat7 = data['origin']
y = data['weight']
cor1 = np.corrcoef(y, feat1)[0,1]
cor2 = np.corrcoef(y, feat2)[0,1]
cor3 = np.corrcoef(y, feat3)[0,1]
cor4 = np.corrcoef(y, feat4)[0,1]
cor5 = np.corrcoef(y, feat5)[0,1]
cor6 = np.corrcoef(y, feat6)[0,1]
cor7 = np.corrcoef(y, feat7)[0,1]
print(cor1)
print(cor2)
print(cor3)
print(cor4)
print(cor5)
print(cor6)
print(cor7)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

y = np.array(data['mpg'])
x = np.array(data['cylinders'])
print(x.shape, y.shape)
reg = linear_model.fit(x[:, None], y)
fit_score = reg.score(x[:, None], y)
#also look at the get_params() command
print(fit_score)
```

```python
#Multiple Linear Regression
#Create a 2D matrix with each column representing different features
#Look up np.concatenate or np.stack
x = np.stack((data['cylinders'], data['weight'], data['horsepower'], data['year'], data['acceleration'], data['origin'], data['displacement']))
print(x.shape)

linear_model = LinearRegression()

y = np.array(data['mpg'])
print(x.shape, y.shape)
reg = linear_model.fit(x.T, y)
fit_score = reg.score(x.T, y)
#also look at the get_params() command
print(fit_score)
```

```python
#Plot the Regression Model
# You will need to make four diagnostic plots
w = reg.coef_
print(w.shape)
pred = x.T @ w
r = (y - pred)
plt.scatter(pred, r)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy import genfromtxt
data1 = genfromtxt('data1.csv', delimiter=',')
X = data1[:,:-1]
y = data1[:, -1]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
projected = pca.fit_transform(X)

plt.scatter(projected[:, 0], projected[:, 1], c=y)
plt.title('First two principal component projections')
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(X)

plt.scatter(projected[:, 0], projected[:, 1], c=clusters)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300)
plt.title('K-means cluster on raw data with K=3')
plt.xlabel('component 1')
plt.ylabel('component 2')
```

*Problem 5.1.b*
The obtained clusters are similar to the true class label.

```python
kmeans = KMeans(n_clusters=2)
clusters = kmeans.fit_predict(X)

plt.scatter(projected[:, 0], projected[:, 1], c=clusters)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300)
plt.title('K-means cluster on raw data with K=2')
plt.xlabel('component 1')
plt.ylabel('component 2')
```

```python
kmeans = KMeans(n_clusters=4)
clusters = kmeans.fit_predict(X)

plt.scatter(projected[:, 0], projected[:, 1], c=clusters)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300)
plt.title('K-means cluster on raw data with K=4')
plt.xlabel('component 1')
plt.ylabel('component 2')
```

*Problem 5.1.d*

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(projected)

plt.scatter(projected[:, 0], projected[:, 1], c=clusters)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300)
plt.title('K-means cluster on first two principal components with K=3')
plt.xlabel('component 1')
plt.ylabel('component 2')
```

```python
# Inter Observational Distance
from scipy.spatial.distance import pdist
interDist = pdist(X, 'euclidean')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z1 = linkage(interDist, 'average')
fig = plt.figure(figsize=(25, 10))
dn1 = dendrogram(Z1)
```

```python
Z2 = linkage(interDist, 'complete')
fig = plt.figure(figsize=(25, 10))
dn2 = dendrogram(Z2)
```

```python
Z3 = linkage(interDist, 'single')
fig = plt.figure(figsize=(25, 10))
dn3 = dendrogram(Z3)
```

```python
fig = plt.figure(figsize=(25, 10))
dn4 = dendrogram(Z1, color_threshold = 12)
```
